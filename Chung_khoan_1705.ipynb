```python
import pandas as pd
import numpy as np
from IPython.display import display
import matplotlib.pyplot as plt
#Read
df=pd.read_csv('Dữ liệu quý chứng khoán năm.csv')
nganh=pd.read_excel('Phân nhóm ngành 21042023.xlsx')
#Merge
df1=pd.merge(df,nganh,left_on='Symbol',right_on='Mã CK')
df1 = df1.rename(columns ={ 'Symbol': 'Mã CK', 'Year': 'Năm','Ngành cấp 1':'Ngành nghề chính','Ngành cấp 3':'Ngành nghề phụ'})
df1
```

Ngành cấp 1 và Ngành cấp 3

```python
#Creat new table
new_table=df1[['Mã CK','Năm','Lưu chuyển tiền thuần từ hoạt động kinh doanh','Lưu chuyển tiền thuần từ hoạt động đầu tư','Lưu chuyển tiền thuần từ hoạt động tài chính','Lưu chuyển tiền thuần trong kỳ','* LNST chưa phân phối lũy kế đến cuối kỳ trước','Tiền và tương đương tiền cuối kỳ','Ngành nghề chính','Ngành nghề phụ']]
new_table
```

Dòng tiền trong năm

```python
import seaborn as sns
dcn1= new_table.groupby(['Ngành nghề chính','Năm'])['Lưu chuyển tiền thuần trong kỳ'].sum().reset_index()
dcn1 = dcn1[dcn1['Năm'] >= 2015]
dcn1 = dcn1[dcn1['Ngành nghề chính'] != 'Tài chính']
plt.figure(figsize=(12, 8))
sns.lineplot(x='Năm', y='Lưu chuyển tiền thuần trong kỳ', hue='Ngành nghề chính', data=dcn1)
plt.title('PHÂN TÍCH BIẾN ĐỘNG DÒNG TIỀN TRONG NĂM CỦA CÁC NGÀNH NGHỀ CHÍNH',fontweight='bold',fontname='Roboto',fontsize=16)
plt.xlabel('Năm')
plt.ylabel('Lưu chuyển tiền thuần trong năm')
plt.legend(loc='upper left')
plt.show()

```

Dòng tiền kinh doanh

```python
import seaborn as sns
dcn1= new_table.groupby(['Ngành nghề chính','Năm'])['Lưu chuyển tiền thuần từ hoạt động kinh doanh'].sum().reset_index()
dcn1 = dcn1[dcn1['Năm'] >= 2015]
dcn1 = dcn1[dcn1['Ngành nghề chính'] != 'Tài chính']
plt.figure(figsize=(12, 8))
sns.barplot(x='Năm', y='Lưu chuyển tiền thuần từ hoạt động kinh doanh', hue='Ngành nghề chính', data=dcn1)
plt.title('PHÂN TÍCH BIẾN ĐỘNG DÒNG TIỀN KINH DOANH CỦA CÁC NGÀNH NGHỀ CHÍNH',fontweight='bold',fontname='Roboto',fontsize=16)
plt.xlabel('Năm')
plt.ylabel('Lưu chuyển tiền thuần từ hoạt động kinh doanh')
plt.legend(loc='upper left')
plt.show()
```

Dòng tiền đầu tư

```python
import seaborn as sns
dcn1= new_table.groupby(['Ngành nghề chính','Năm'])['Lưu chuyển tiền thuần từ hoạt động đầu tư'].sum().reset_index()
dcn1 = dcn1[dcn1['Năm'] >= 2015]
dcn1 = dcn1[dcn1['Ngành nghề chính'] != 'Tài chính']
plt.figure(figsize=(12, 8))
sns.barplot(x='Năm', y='Lưu chuyển tiền thuần từ hoạt động đầu tư', hue='Ngành nghề chính', data=dcn1)
plt.title('PHÂN TÍCH BIẾN ĐỘNG DÒNG TIỀN ĐẦU TƯ CỦA CÁC NGÀNH NGHỀ CHÍNH',fontweight='bold',fontname='Roboto',fontsize=16)
plt.xlabel('Năm')
plt.ylabel('Lưu chuyển tiền thuần từ hoạt động đầu tư')
plt.legend(loc='lower left',fontsize=9)
plt.show()
```
